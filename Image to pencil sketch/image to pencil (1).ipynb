```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
from IPython.display import display
print("Necessary packages included successfully!")
```

```python
img=cv2.imread("Tom and jerry.png")
```

```python
img=cv2.resize(img,(700,500))
```

```python
img
```

```python
img_rgb=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
display(Image.fromarray(img))
```

```python
gray_img=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
gray_img
```

```python
display(Image.fromarray(gray_img))
```

```python
inverted_img=cv2.bitwise_not(gray_img)
```

```python
inverted_img
```

```python
display(Image.fromarray(inverted_img))
```

```python
img_smoothing=cv2.GaussianBlur(inverted_img,(21,21),sigmaX=0,sigmaY=0)
```

```python
img_smoothing
```

```python
display(Image.fromarray(img_smoothing))
```

```python
def dodgev2(x,y):
    return cv2.divide(x,255-y,scale=256.0)
```

```python
pencil_sketch=dodgev2(gray_img,img_smoothing)
```

```python
display(Image.fromarray(pencil_sketch))
```

```python
display(Image.fromarray(img_rgb))
display(Image.fromarray(pencil_sketch))
```
